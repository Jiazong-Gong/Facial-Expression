```python
import cv2
import json
from matplotlib import pyplot as plt
import seaborn as sns
```

```python
file = open('./face_data.json')
file = json.load(file)
file
```

```python
classes_to_index = {"Angry": 0, "Disgust": 1, "Fear": 2, "Happy": 3, "Sad": 4, "Surprise": 5, "Neutral": 6}
idx_to_classes = {}
for (k, v) in classes_to_index.items():
    idx_to_classes[v] = k
```

```python
idx_to_classes[0]
```

```python
examples = []
examples.append(file['IAmSam_004235880_00000064'])
examples.append(file['HarryPotter_Deathly_Hallows_1_011317280_00000001'])
examples.append(file['HarryPotter_Half_Blood_Prince_005938934_00000045'])
examples.append(file['MissMarch_000546640_00000065'])
```

```python
counter = 0
plt.figure(figsize=(5, 6))
plt.title('Some Examples')
for i in examples:
    counter += 1
    image_path = i['Image']
    image = cv2.imread(image_path)
    class_idx = i['Class']
    class_name = idx_to_classes[class_idx]
    plt.subplot(2, 2, counter), plt.imshow(image), plt.title(class_name)
```
